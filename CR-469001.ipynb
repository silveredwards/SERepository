# market definition

```python
%%read_sql

---- aim to create market def sheet in output file
create or replace transient table hw_CR_469001_market_def as (
    select 'ICD10' as code_type, 
            level_4 as code, 
            level_4_short_description_icd10 as description
    from RWD_DB.RWD_REFERENCE_LIBRARY.ICD_GROUPER
    where level_4 ilike 'M05%'
        or level_4 ilike 'M06%'
        or level_4 ilike 'L40%'
    union all
    select 'ICD9'as code_type,
            ICD9_MAPPED_CODES as code, 
            ICD9_DESCRIPTION as description
    from RWD_DB.RWD_REFERENCE_LIBRARY.ICD_GROUPER
    where ICD9_MAPPED_CODES ilike '696%'
        or ICD9_MAPPED_CODES ilike '714%'

);
```

# pull data from dx table

```python
%%read_sql

---- intitial action: pull data from diagnosis by dx code given in market def table
---- result: check if using market def from ref table would lead missing dx codes
---- for details, to see QC part
---- it does miss some values
---- action: aplly ilike function to filter data on dx table directly



create or replace transient table hw_CR_469001_data_pull as (
       SELECT 
        dx.patient_id,
        dx.claim_id,
        dx.code_type, 
        dx.diagnosis_sequence,
        dx.year_of_service,
        dx.diagnosis,
        dx.etl_create_ts,
        dx.claim_type_code
    FROM rwd_db.rwd.raven_claims_submits_diagnosis dx
    where  dx.diagnosis in (select code from hw_CR_469001_market_def)
        and year_of_service BETWEEN '{start_date}' AND '{end_date}'
        and patient_id is not null
        and claim_id is not null
        and data_source <> 'VULTURE'
        and data_source <> 'PUFFIN'
);
```

```python
%%read_sql

show tables like 'hw_CR_469001_data_pull'
```

```python
%%read_sql

--add  dx name
create or replace transient table hw_CR_469001_data_pull_2 as (
        select a.*,
                b.description,
                case
                    when a.diagnosis ilike '714%' or a.diagnosis ilike 'M05%' or a.diagnosis ilike 'M06%'
                        then 'RA'
                    when a.diagnosis ilike '696%' or a.diagnosis ilike 'L40%'
                        then 'PSO'
                end as disease
        from hw_CR_469001_data_pull a
        left join 
        hw_CR_469001_market_def b
        on a.diagnosis = b.code
)
```

# counts

```python
%%read_sql

---- make counts
select year(year_of_service) as year, 
        disease,
        count(distinct  patient_id) as patient_vol,
        count(distinct claim_id) as claim_vol
from hw_CR_469001_data_pull_2 a
group by rollup (1,2)
order by 1,2;
```

```python
%%read_sql

---- make counts only by year

select year(year_of_service) as year, 
        count(distinct  patient_id) as patient_vol,
        count(distinct claim_id) as claim_vol
from hw_CR_469001_data_pull_2 a
group by rollup (1)
order by 1;
```

```python
%%read_sql

---- make counts of each disease
select  'RA overall' as group_name,
        count(distinct  patient_id) as patient_vol,
        count(distinct claim_id) as claim_vol
from hw_CR_469001_data_pull_2 a
where disease = 'RA'
union all
select  'PSO overall' as group_name,
        count(distinct  patient_id) as patient_vol,
        count(distinct claim_id) as claim_vol
from hw_CR_469001_data_pull_2 a
where disease = 'PSO'
union all
select  'overall' as group_name,
        count(distinct  patient_id) as patient_vol,
        count(distinct claim_id) as claim_vol
from hw_CR_469001_data_pull_2 a;
```

# QC part


```python
%%read_sql

--pull data from diagnosis by dx code given in market def table 

create or replace transient table hw_CR_469001_data_pull_check as (
       SELECT 
        dx.patient_id,
        dx.claim_id,
        dx.code_type, 
        dx.diagnosis_sequence,
        dx.year_of_service,
        dx.diagnosis,
        dx.etl_create_ts,
        dx.claim_type_code
    FROM rwd_db.rwd.raven_claims_submits_diagnosis dx
    where diagnosis in (select code from hw_CR_469001_market_def)
        and year_of_service BETWEEN '{start_date}' AND '{end_date}'
        and patient_id is not null
        and claim_id is not null
        and data_source <> 'VULTURE'
        and data_source <> 'PUFFIN'
);
```

```python
%%read_sql

---- year-wisely analysis

select year(year_of_service) as year, 
        count(distinct patient_id) as patient_vol,
        count(distinct claim_id) as claim_vol
from hw_CR_469001_data_pull_check 
group by rollup (1)
order by 1;
```

```python
%%read_sql

---- dx code compare
---- check and update ICD code into market def step
with market_def_code as (
        select distinct diagnosis as market_dx
        from hw_CR_469001_data_pull_check 
        order by 1),
        
actual_code as (
        select distinct diagnosis as actual_dx
        from hw_CR_469001_data_pull_2
        order by 1)
select actual_dx,market_dx
from actual_code a
full outer join market_def_code b
on a.actual_dx = b.market_dx
order by 1;
```
